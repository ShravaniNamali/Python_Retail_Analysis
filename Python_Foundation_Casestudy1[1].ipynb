### 11. For all customers aged between 25 - 35, find out:
### a. What was the total amount spent for “Electronics” and “Books” product categories?
### b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?

```python
from datetime import datetime
```

```python
(datetime.today() - Customer_Final.DOB)
```

```python
Age = Customer_Final.DOB.apply(lambda x: (datetime.today() - x).days) /365
```

```python
Customer_Final['Age'] = Age
```

```python
Aged = Customer_Final.loc[(Customer_Final['Age'] > 25) & (Customer_Final['Age'] < 35)]
```

```python
Aged
```

```python
# a. What was the total amount spent for “Electronics” and “Books” product categories?¶

Books_Electronics = Aged.loc[Aged['prod_cat'].isin(['Electronics','Books'])]
```

```python
Books_Electronics.groupby('prod_cat')['total_amt'].sum()
```

```python
# b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?

Start = pd.to_datetime("2014-01-01")
End = pd.to_datetime("2014-03-01")
```

```python
Start_End = Aged.loc[((Aged['tran_date'] > Start) & (Aged['tran_date'] < End))]
```

```python
Start_End.total_amt.sum()
```
